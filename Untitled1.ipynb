## Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Importing and creating the datasets

```python
data_2019=pd.read_excel("Data Sheet.xlsx","Data 2019")
data_2019
```

```python
data_2020=pd.read_excel("Data Sheet.xlsx","Data 2020")
data_2020
```

## Data preprocessing

### 1. Removing the unwanted data

```python
data_2020.drop(["Unnamed: 0","CPU"],axis=1,inplace=True)
```

```python
data_2019.drop(["Unnamed: 0","Unnamed: 1","CPU"],axis=1,inplace=True)
```

```python
data_2020=data_2020.drop(data_2020[(data_2020['Operating System']!='IOS') & 
                                   (data_2020['Operating System']!='Android')].index)
```

```python
data_2019=data_2019.drop(data_2019[(data_2019['Operating System']!='IOS') & 
                                   (data_2019['Operating System']!='Android')].index)
```

```python
data_2019.dropna(inplace=True)
```

```python
data_2020.dropna(inplace=True)
```

### 2. Dealling with categorical data

```python
brands=data_2020['Brand Name'].unique()
brands
```

```python
l=[5,2,3,5,2,2,2,1,1,2,4,2,2,4,3,5,3,3,5,2,3,3,3,1]
```

```python
brand=pd.Series(l,index=brands)
```

```python
data_2020
```

```python
data_2020["Brand Value"]=brand[data_2020["Brand Name"]].values
```

```python
data_2020
```

```python
brands=data_2019['Brand Name'].unique()
brands
```

```python
l=[5,2,3,5,2,2,2,2,2,2,4,3,5,3,3,5,2,3,3,1]
```

```python
brand=pd.Series(l,index=brands)
```

```python
data_2019["Brand Value"]=brand[data_2019["Brand Name"]].values
```

### 3. Add a column

```python
data_2019["Over_ALL"]=data_2019["Bench Score"]+5000*data_2019["Brand Value"]+1000*data_2019["Camera Quality"]+100*1/data_2019[
    "Price"]+1000*data_2019["RAM"]
```

```python
data_2020["Over_ALL"]=data_2020["Bench Score"]+5000*data_2020["Brand Value"]+100*1/data_2020[
    "Price"]+1000*data_2020["RAM"]+1000*data_2020["Camera Quality"]
```

### 4. Defining the Budget segment

```python
budget2019=data_2019[(data_2019["Price"]<=20000) & (data_2019["Price"]>=10000)] 
budget2019
```

```python
budget2020=data_2020[(data_2020["Price"]<=20000) & (data_2020["Price"]>=10000)] 
budget2020
```

```python
budget2019.sort_values(by=["Over_ALL"],ascending=False)#[["Brand Name","Model No","Price"]].head()
```

```python
budget2020.sort_values(by=["Over_ALL"],ascending=False)[["Brand Name","Model no","Price"]].head()
```
